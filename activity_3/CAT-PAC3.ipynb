## Exercici 1
El número $\pi$ es pot aproximar amb la següent sèrie de sumands:

    pi = 4 (1-1/3+1/5-1/5+1/7-1/7+1/9...)

només cal donar més valors a la sèrie, i obtindrem una millor aproximació. Aquesta sèrie la podem implementar amb la següent funció:

    def approximate_pi(N):
        value = 0
        for k in range(1, 2*N+1, 2):
            sign = -(k % 4 - 2)
            value += float(sign) / k
        return 4 * value

<span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> **(1 punt)**

### 1.1
Compta els passos que realitza la funció **approximate_pi()** i dedueix-ne la seva complexitat assimptòtica en el pitjor cas (notació $\mathcal{O}$), a partir del número de passos que s'executen.

**Solució 1.1**.
Aproximadament aquest algoritme té complexitat **O(N)** perque el nombre d'operacions que fem al bucle són 2N+1-1 = 2N pero el step=2 per tant la meitat que és N. La operació módul té un cost despreciable i l'operació més costosa és "value += float(sign) / k" que és realitza N cops pero tenen temps d'execució constant (ja que no depenen de la llargaria de N). El return seria una operació extra ja que multipliquem per 4 pero, a nivell asimptotic no afecta perque multipliquem per una constant, llavors és N.



### 1.2
Confirma aquest resultat fent una gràfica del temps que triga a executar-se l'algorisme amb diferents valors de número de sumands. Pots provar fins a 10000000.

**Solució 1.2**

```python
import time
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


# Definim la funció:
def approximate_pi(N):
    """
    Aquesta funció aproxima pi per una sèrie convergent 
    """
    value = 0
    for k in range(1, 2*N+1, 2):
        sign = -(k % 4 - 2)
        value += float(sign) / k
    return 4 * value
 

# Ara calculem valors per tenir la gràfica:
def time_pi_funct(n):
    """
    Aquesta funció calcula el temps d'execució de la 
    funció approximate_pi, retornant-ne el temps 
    en segons.
    """
    init = time.time()
    approximate_pi(n)
    final = time.time() - init
    return final


# Creem la funció per fer el plot:
def plot_funct(xs):
    """
    Introduim un range per partir el espai de x
    i calcularà el gràfic de temps per nombre 
    d'iteracions (calculant potències de 10).
    """
    # Creem la llista de punts
    xvalues = [10**i for i in xs]
    yvalues = [time_pi_funct(n) for n in xvalues]
    
    # Fem el plot
    plt.figure()
    plt.plot(xvalues, yvalues, '-', color='tab:brown')
    plt.legend(["aprox_pi_funct"])
    plt.xlabel('$n (nombre d\'iteracions)$')
    plt.ylabel('$t(n) - segons)$')
    plt.title("Temps d'execució - test")
    plt.show()


# Cridem la funció:
plot_funct(range(0,9))

# El comportament és clarament lineal
```

Els dos primers, com el seu nom indica, són mètodes lineals, que vol dir que el temps de cerca serà proporcional al número de valors a la llista (ens situem al pitjor cas de què cerquem l'últim valor de la llista).

### 3.1
Implementa els mètodes **Linear Search** i **Sentinel Linear Search**. Un número que segur que existeix és el primer de la llista, que ja l'has trobat. Un número que segur que no existeix és qualsevol número de 7 xifres. Amb quatre *assert* (2 per cada mètode) comprova aquests resultats.

**NOTA:** El mètode *Sentinel Linear Search* no ha de retornar un print, ha de retornar l'index de la posició on s'ha trobat el valor.

**NOTA:** a l'hora de copiar/enganxar codi d'una pàgina web, s'ha d'anar amb compte amb com es copien les identacions i tabuladors.

**Solució 3.1**

```python
# Pel métode lineal:
def linealSearch(numToLookUp):
    """Retornarà el index de la llista o -1 en cas que no trobi res"""

    num = 0
    while num <len(llista_brut_sense_rep):
        if llista_brut_sense_rep[num] == numToLookUp:
            return num
        num += 1
    return -1

# Pel métode Sentinel Lineal:
def sentinelSearch(numToLookUp):
    mida = len(llista_brut_sense_rep)
    """
        Faig notar que el no precalcular el len de la llista fa 
        que tecnicament hi haguin més operacions pero 
        son irrellevants per calcular
        la complexitat
    """
 
    # Guardarem el últim element de la llista
    lastElement = llista_brut_sense_rep[mida - 1]
 
    # Fiquem el element a buscar conm ultim element
    llista_brut_sense_rep[mida - 1] = numToLookUp
    
    # Fiquem el índex
    index = 0
     
    # Ho faig amb while perque amb aquest mètode no ens surtirem 
    # del rang i acabarem en temps finit.
    while (llista_brut_sense_rep[index] != numToLookUp):
        index += 1
 
    # Put the last element back
    llista_brut_sense_rep[mida - 1] = lastElement
 
    if ((index < mida - 1) or (llista_brut_sense_rep[mida - 1] == numToLookUp)):
        return index
    else:
        return -1
```

```python
# asserts de la lineal:
assert linealSearch(first_element) == 0, "No hem trobat el primer element"
```

```python
assert linealSearch(1000000) == -1, "hi ha elements de mes de 6 digits"
```

```python
# asserts de la sentinel:
assert sentinelSearch(first_element) == 0,  "no hem trobat el primer element"
```

```python
assert sentinelSearch(1000000) == -1, "hi ha elements de mes de 6 digits"
```

```python
# Definim les funcions per calcular el temps d'execució:
# Pel mètode lineal
def time_linear_funct(element_to_search):
    """
    calcula el temps d'execució
    de la funció de cerca lineal
    """
    init = time.time()
    linealSearch(element_to_search)
    final = time.time() - init
    return final

# Pel mètode sentinel
def time_sentinel_funct(element_to_search):
    """
    calcula el temps d'execució
    de la funció de cerca sentinel
    """
    init = time.time()
    sentinelSearch(element_to_search)
    final = time.time() - init
    return final
```

Implementa els tres primers algorismes de la llista: Binary Search, Meta Binary Search i Ternary Search. Posa en una mateixa gràfica per tal d'observar si hi ha alguna diferència significativa. Recorda que aquests algorismes són de tipus log(N), i per tant si poses escala logarítmica en el temps d'execució hauries de veure un comportament lineal.

**Solució 4.2**

```python
# Implementem els 3 algoritmes (binary ja està implementat):
# Meta Binary:
import math  # Aquest codi i el de terciary son de la URL (he fet modificacions menors) NO ES MEU
def metaBinarySearch(key_to_search):
 
    n = len(llista_brut_sense_rep)
    # El nombre de bits (li dirè log_discretitzat perque és el exponent de 2^(n-1) fent log base 2)
    log_discretitzat = int(math.log2(n-1)) + 1;
 
    pos = 0
    # Ens movem del log fins a -1 en salts de -1
    for i in range(log_discretitzat - 1, -1, -1) :
        if (llista_brut_sense_rep[pos] == key_to_search):
            return pos

        # Incrementally construct the
        # index of the target value
        new_pos = pos | (1 << i)

        # ens movem a la nova poss.
        if ((new_pos < n) and (llista_brut_sense_rep[new_pos] <= key_to_search)):
            pos = new_pos
 
    # if element found return
    # pos otherwise -1
    return (pos if(llista_brut_sense_rep[pos] == key_to_search) else -1)

# El métode per calcular el temps del meta:
def time_meta_funct (key_to_search):
    init = time.time()
    metaBinarySearch(key_to_search)
    final = time.time() - init
    return final


# Ternary: (el codi és del link proporcionart NO ES MEU)
# Python3 program to illustrate
# recursive approach to ternary search
import math as mt
# l la poss inicial del array r la final
# son left and right, key el valor buscat i
# ar l'array o llista en aquest cas.


# Function to perform Ternary Search
def ternarySearch(l, r, key, ar):

	if (r >= l):

		# Find the mid1 and mid2
		mid1 = l + (r - l) //3
		mid2 = r - (r - l) //3

		# Check if key is present at any mid
		if (ar[mid1] == key): 
			return mid1
		
		if (ar[mid2] == key): 
			return mid2
		
		# Since key is not present at mid,
		# check in which region it is present
		# then repeat the Search operation
		# in that region
		if (key < ar[mid1]): 

			# The key lies in between l and mid1
			return ternarySearch(l, mid1 - 1, key, ar)
		
		elif (key > ar[mid2]): 

			# The key lies in between mid2 and r
			return ternarySearch(mid2 + 1, r, key, ar)
		
		else: 

			# The key lies in between mid1 and mid2
			return ternarySearch(mid1 + 1, 
								mid2 - 1, key, ar)
		
	# Key not found
	return -1


# El métode per calcular el temps del ternari:
def time_ternary_funct (initial_poss, length, key_to_search, array):
    init = time.time()
    ternarySearch(initial_poss, length, key_to_search, array)
    final = time.time() - init
    return final
```

```python
import decimal
import math


# Aquest codi es el de la url NO ES MEU.
def compute_pi(n):
  """
  This function calculates the value of pi to 'n' number of decimal places
  Args:
  n:   precision(Decimal places)
  Returns:
  pi:   the value of pi to n-decimal places
  """
  
  decimal.getcontext().prec = n + 3
  decimal.getcontext().Emax = 999999999
  
  C = 426880 * decimal.Decimal(10005).sqrt()
  K = decimal.Decimal(6)
  M = decimal.Decimal(1)
  X = decimal.Decimal(1)
  L = decimal.Decimal(13591409)
  S = L
  
  # For better precision, we calculate to n+3 and truncate the last two digits
  for i in range(1, n+3):
    M = decimal.Decimal(M* ((1728*i*i*i)-(2592*i*i)+(1104*i)-120)/(i*i*i))
    L = decimal.Decimal(545140134+L)
    X = decimal.Decimal(-262537412640768000*X)
    S += decimal.Decimal((M*L) / X)
    
  return str(C/S)[:-2] # Pi is C/S

# Guardem els digits en una variable.
pi_with_100_decimals = compute_pi(100)

# Imprimim la variable:
print(pi_with_100_decimals)
```
