````python
# Imports:
import os
import zipfile as zf
# from sys import getsizeof
import glob
import re


# Variables:
os.chdir("/content/")
print(os.getcwd())
PATH_ORIGINAL_ZIP = "/content/drive/MyDrive/Colab Notebooks/activity_2/original.zip"
PATH_TO_LOOKUP = "./original"

# Exercici 2.1 ---------
# Descomprimim el fitxer per codi

with zf.ZipFile(PATH_ORIGINAL_ZIP, 'r') as topsecret:
    os.chdir("/content/")
    topsecret.extractall()
    print("Hem descomprimit el arxiu")

# Tornem al nostre dir
print(os.getcwd())
os.chdir("./drive/MyDrive/Colab Notebooks/activity_2/")
# Hem de descomprimir el fitxer per codi de ciutat


def sizeLookUp(path, conversion_to=1024**2):
    """
        <string> path ----> funt sizeLookUp ----> <int> size
            Aquesta func. entre un <string> (path) i actua de dues
            formes depenent de si el path es de un dir o un archiu:
                - directori ---> retorna mida de tota la carpeta
                - ficher ------> retorna la mida del fixer
    """
    total = 0
    # Comprovem si ens trobem en un directori
    if os.path.isdir(path):
        # Llegim tots els documents del directori
        with os.scandir(path) as folder:
            # si no hi ha elements al dir pasara
            # el for sense fer res i retornarà el total
            # que serà zero per un dir buit
            for folder_element in folder:
                # apliquem recursivament la funció
                total += sizeLookUp(folder_element.path)
    # Si tenim un fitxer, lleguim el tamany del arxiu
    else:
        total += os.path.getsize(path)/conversion_to
        # Deixo aquesta primera versió de la funció
        # com a fe d'errates ja que, tenia discrepàncies amb
        # els valors correctes. Ho he solucionat amb os.path.getsize
        # with open(path, 'r') as file:
        #     total += getsizeof(file.read())/(conversion_to)
    # Fem la conversió
    return total


# Guardem per pasar els test la info
size_original = sizeLookUp(PATH_ORIGINAL_ZIP)
size_decompressed = sizeLookUp("/content/original")

# Mirem la mida del arxiu sense descomprimir
actual_exercice = "2.0"

# incrementem el exercici
actual_exercice = str(float(actual_exercice)+0.1)
print("Per el exercici {}:".format(actual_exercice))
print("\tEl path on mirem la mida del .zip és {}".format(PATH_ORIGINAL_ZIP))
print("\t\t{:.2f} MB".format(size_original))

# Executem la funció per la següent ruta
print("\tEl path on mirem la mida és {}".format(PATH_TO_LOOKUP))
print("\t\t{:.2f} MB".format(size_decompressed))
````

````python
# Exercici 3.2 -----------


import re
# Patrons que busquem

reg1 = r'\b[Uu]\w*\s+[Ff]\w*\s+[Oo]\w*'
reg2 = r'\bE\w*\s+T\w*'

# alternativament E\w*\s+T\w*
# Fem una reserca mès brusca perque nomès volem veure
# si tenim el match, per tant els reg finals son mès curts
# Eliminarem les rows que no continguin els dos patrons anteriors


def verifyConditions(text):
    """
    Funció booleana (retorna un boolea) rep un index
    i comprova si es verifiquen dos reggex
        en cas afirmatiu: true
        en cas contrari false
    """
    global reg1
    global reg2

    # 7 puede dar problema pilla la palabra  // 103 ejemplo bueno
    if re.search(reg1, text) and re.search(reg2, text):
        return True
    return False


# Eliminem les possicions on la nostra funció retorni false
df_ex3_2 = df.loc[df["text"].apply(verifyConditions)]
````
